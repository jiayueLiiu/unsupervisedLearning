```python
import torchvision
import torchvision.transforms as transforms

#############################################
# your code here
# specify preprocessing function
transform_train = transforms.Compose([
    # transforms.RandomCrop(32, 4),
    # transforms.RandomHorizontalFlip(),
    transforms.ToTensor(),
    transforms.Normalize([0.4914, 0.4822, 0.4465], [0.2023, 0.1994, 0.2010]),
    
])

transform_val = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize([0.4914, 0.4822, 0.4465], [0.2023, 0.1994, 0.2010])
])
```

```python
from tools.dataset import CIFAR10
from torch.utils.data import DataLoader

# a few arguments, do NOT change these
DATA_ROOT = "./data"

"""
Here is where we define batch size
"""

TRAIN_BATCH_SIZE = 128 
VAL_BATCH_SIZE = 250

#############################################
# your code here
# construct dataset
train_set = CIFAR10(
    root=DATA_ROOT, 
    mode='train', 
    download=True,
    transform = transform_train    # your code
)
val_set = CIFAR10(
    root=DATA_ROOT, 
    mode='val', 
    download=True,
    transform= transform_val    # your code
)

# construct dataloader
train_loader = DataLoader(
    train_set, 
    batch_size=TRAIN_BATCH_SIZE,  # your code
    shuffle=True,     # your code
    num_workers=4
)
val_loader = DataLoader(
    val_set, 
    batch_size=VAL_BATCH_SIZE,  # your code
    shuffle=True,     # your code
    num_workers=4
)
```

```python

INITIAL_LR = 0.1

# momentum for optimizer
MOMENTUM = 0.9

# L2 regularization strength
REG = 5e-4

#############################################
# your code here
# create loss function
criterion = nn.CrossEntropyLoss()

# Add optimizer
#optimizer = optim.Adam(net.parameters(), lr = 0.001,weight_decay = 0)
optimizer = optim.SGD(net.parameters(), lr = INITIAL_LR, momentum = MOMENTUM, weight_decay = REG)
```

```python
#do not run this block if rotNet is trained
import matplotlib.pyplot as plt
import random
# some hyperparameters
# total number of training epochs
EPOCHS = 100

# the folder where the trained model is saved
CHECKPOINT_FOLDER = "./saved_model"

# start the training/validation process
# the process should take about 5 minutes on a GTX 1070-Ti
# if the code is written efficiently.
best_val_acc = 0
# current_learning_rate = INITIAL_LR

current_learning_rate = 0.1

DECAY_EPOCHS = 30
DECAY = 0.2
print("==> Training starts!")
print("="*50)

for param_group in optimizer.param_groups:
    param_group['lr'] = current_learning_rate

for i in range(0, EPOCHS):
    
    # handle the learning rate scheduler.
    if i % DECAY_EPOCHS == 0 and i > 10:
        current_learning_rate = current_learning_rate * DECAY
        for param_group in optimizer.param_groups:
            param_group['lr'] = current_learning_rate
        print("Current learning rate has decayed to %f" %current_learning_rate)
    
    #######################
    # your code here
    # switch to train mode
    net.train(True)
    #######################
    
    print("Epoch %d:" %i)
    # this help you compute the training accuracy
    total_examples = 0
    correct_examples = 0

    train_loss = 0 # track training loss if you want
    
    # Train the model for 1 epoch.
    for batch_idx, (inputs, targets) in enumerate(train_loader):
        ####################################
        # your code here
        # copy inputs to device
        # print(inputs.shape)
        batch_inputs = np.zeros([inputs.shape[0]*4,inputs.shape[1],inputs.shape[2],inputs.shape[3]])
        # batch_inputs = []
        labels = np.zeros([targets.shape[0]*4])
        batch_ind = 0
        
        for image in inputs:
          image_rotated0 = image
          image_rotated1 = np.rot90(image_rotated0,axes = (1,2))
          image_rotated2 = np.rot90(image_rotated1,axes = (1,2))
          image_rotated3 = np.rot90(image_rotated2,axes = (1,2))

          images = [image_rotated0, image_rotated1, image_rotated2, image_rotated3]
          label = [0, 1, 2, 3]
          indices = list(range(len(images)))


          batch_inputs[batch_ind*4,:,:,:] = image_rotated0
          batch_inputs[batch_ind*4+1,:,:,:] = image_rotated1
          batch_inputs[batch_ind*4+2,:,:,:] = image_rotated2
          batch_inputs[batch_ind*4+3,:,:,:] = image_rotated3


          labels[batch_ind*4] = 0
          labels[batch_ind*4+1] = 1
          labels[batch_ind*4+2] = 2
          labels[batch_ind*4+3] = 3

          batch_ind +=1
        

        indices = np.arange(batch_inputs.shape[0])
        np.random.shuffle(indices)
        batch_inputs = batch_inputs[indices,:,:,:]
        labels = labels[indices]

        
        # print(torch.Tensor(batch_inputs).shape)
        batch_inputs = torch.Tensor(batch_inputs) 
        labels = torch.Tensor(labels) 
        labels = labels.type(torch.LongTensor)
        # print(torch.Tensor(labels).shape)

        # batch_size, rotations, channels, height, width = batch_inputs.size()
        batch_inputs = batch_inputs.to(device)
        labels = labels.to(device)

        out, _ = net.forward(batch_inputs)
        loss = criterion(out, labels)
        
        # inputs = inputs.to(device)
        # labels = targets.to(device)
        # # compute the output and loss
        # _, out = net.forward(inputs)
        # loss = criterion(out, labels)
        train_loss += loss.item()
        #print(loss.item())
        # zero the gradient
        optimizer.zero_grad()
        
        # backpropagation
        loss.backward()   
        
        # apply gradient and update the weights
        optimizer.step()
        
        # count the number of correctly predicted samples in the current batch
        _, predicted = torch.max(out, 1)
        # ############
        correct_examples = (predicted == labels).sum().detach().cpu().numpy()
        # correct_examples = (predicted == targets).sum().detach().cpu().numpy()
        # print((predicted == targets).sum().detach().cpu().numpy())
        total_examples = len(labels)
        ####################################
                
    avg_loss = train_loss / len(train_loader)
    #print(avg_loss)
    avg_acc = correct_examples / total_examples
    print("Training loss: %.4f, Training accuracy: %.4f" %(avg_loss, avg_acc))

    # Validate on the validation dataset
    #######################
    # your code here
    # switch to eval mode
    net.train(False)
    
    #######################

    # this help you compute the validation accuracy
    total_examples = 0
    correct_examples = 0
    
    val_loss = 0 # again, track the validation loss if you want

    # disable gradient during validation, which can save GPU memory
    with torch.no_grad():
        for batch_idx, (inputs, targets) in enumerate(val_loader):


          batch_inputs = np.zeros([inputs.shape[0]*4,inputs.shape[1],inputs.shape[2],inputs.shape[3]])
          # batch_inputs = []
          labels = np.zeros([targets.shape[0]*4])
          batch_ind = 0
          for image in inputs:
            image_rotated0 = image
            image_rotated1 = np.rot90(image_rotated0,axes = (1,2))
            image_rotated2 = np.rot90(image_rotated1,axes = (1,2))
            image_rotated3 = np.rot90(image_rotated2,axes = (1,2))

            batch_inputs[batch_ind*4,:,:,:] = image_rotated0
            batch_inputs[batch_ind*4+1,:,:,:] = image_rotated1
            batch_inputs[batch_ind*4+2,:,:,:] = image_rotated2
            batch_inputs[batch_ind*4+3,:,:,:] = image_rotated3
            

            labels[batch_ind*4] = 0
            labels[batch_ind*4+1] = 1
            labels[batch_ind*4+2] = 2
            labels[batch_ind*4+3] = 3

            batch_ind +=1
          
          # print(torch.Tensor(batch_inputs).shape)
          batch_inputs = torch.Tensor(batch_inputs) 
          labels = torch.Tensor(labels) 
          labels = labels.type(torch.LongTensor)
          # print(torch.Tensor(labels).shape)
          indices = np.arange(batch_inputs.shape[0])
          np.random.shuffle(indices)
          batch_inputs = batch_inputs[indices,:,:,:]
          labels = labels[indices]

          # batch_size, rotations, channels, height, width = batch_inputs.size()
          batch_inputs = batch_inputs.to(device)
          labels = labels.to(device)
          # compute the output and loss
          out,_ = net(batch_inputs)
          loss = criterion(out, labels)
          
          # inputs = inputs.to(device)
          # labels = targets.to(device)
          # # compute the output and loss
          # _, out = net.forward(inputs)
          # loss = criterion(out, labels)

          val_loss +=loss.item()
          # count the n_, predicted = torch.max(outputs.data, 1)umber of correctly predicted samples in the current batch
          _, predicted = torch.max(out.data, 1)
          total_examples = len(labels)
          correct_examples=(predicted == labels).sum().detach().cpu().numpy()
          
          ###################################

    avg_loss = val_loss / len(val_loader)
    avg_acc = correct_examples / total_examples
    print("Validation loss: %.4f, Validation accuracy: %.4f" % (avg_loss, avg_acc))
    
    # save the model checkpoint
    if avg_acc > best_val_acc:
        best_val_acc = avg_acc
        if not os.path.exists(CHECKPOINT_FOLDER):
           os.makedirs(CHECKPOINT_FOLDER)
        print("Saving ...")
        state = {'state_dict': net.state_dict(),
                'epoch': i,
                'lr': current_learning_rate}
        torch.save(state, os.path.join(CHECKPOINT_FOLDER, 'rotNet_10_resnet.pth'))
        
    print('')

print("="*50)
print(f"==> Optimization finished! Best validation accuracy: {best_val_acc:.4f}")
```

```python
class SimpleNet(nn.Module):
    def __init__(self, input_dim=512, num_classes=10):
        super(SimpleNet, self).__init__()
        self.input_dim = input_dim
        self.num_classes = num_classes
        self.dropout = nn.Dropout(0.85)
        self.fc1 = nn.Linear(self.input_dim, self.num_classes)
        

    def forward(self, x):
        x = self.fc1(x)
        x = self.dropout(x)
        x = F.relu(x)
        return x
```

```python
def train_validate(model, loader, optimizer , is_train, epoch, use_cuda):
    
    loss_func = nn.CrossEntropyLoss()

    data_loader = loader
    # loader.train_loader if is_train else loader.test_loader

    model.train() if is_train else model.eval()
    desc = 'Train' if is_train else 'Validation'

    total_loss = 0
    total_acc = 0
    
    # tqdm_bar = tqdm(data_loader)
    for batch_idx, (x, y) in enumerate(data_loader):
        batch_loss = 0
        batch_acc = 0

        x = x.cuda() if use_cuda else x
        y = y.cuda() if use_cuda else y

        # Get features
        _, outSup = model(x)
        # f_x = f_x.detach()
        # #print(f_x[0])
        # # print(f_x.shape)
        # # Classify features
        # y_hat = classifier_model(f_x)
        # y_hat = feature_model(x)
        # print(y_hat)
        loss = loss_func(outSup, y)

        if is_train:
            # classifier_model.zero_grad()
            optimizer.zero_grad()
            # optimizer_feature.zero_grad()
            loss.backward()
            # optimizer_classifier.step()
            optimizer.step()

        # Reporting
        batch_loss = loss.item() / x.size(0)
        total_loss += loss.item()
        pred = outSup.max(dim=1)[1]
        # print(y_hat.shape)
        # print(y)
        correct = pred.eq(y).sum().item()
        correct /= y.size(0)
        batch_acc = (correct * 100)
        total_acc += batch_acc
        
        # tqdm_bar.set_description('{} Epoch: [{}] Batch Loss: {:.4f} Batch Acc: {:.4f}'.format(desc, epoch, batch_loss, batch_acc))
    print('loss = '+str(total_loss))
    print('Epoch = '+str(epoch) + '; acc = ' + str(total_acc/(batch_idx + 1)))
```

```python

for epoch in range(40):
    train_validate(model, train_loader_classifier, optimizer, True, epoch, True)
    print('========test =====')
    train_validate(model, test_loader_classifier, optimizer, False, 1, True)
    print('==================')
    print('==================')
    

```
